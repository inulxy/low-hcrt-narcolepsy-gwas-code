# Step 2: Send to imputation
1. Download all the *Genotype_QC_Package_NIH_TOPMed/After_QC/pre_impute_FILTERED*.vcf.gz on your local computer
2. Create and then login to TOPMed Imputation Server (https://imputation.biodatacatalyst.nhlbi.nih.gov) using your institution email.
3. Click "Run" on the top of the page and then Genotype Imputation (Minimac4)
4. Upload the 22 autosomes for one ancestry

5. Select options:
- Reference Panel: TOPMed R3
- Array Build: GRCh37/hg19
- rsq Filter: 0.3
- Phasing: Eagle v2.4 (phased output)
- Population: vs TOPMed Panel
- Mode: Quality Control & Imputation
- Select both checkboxes and submit jobs!
- Repeat for each ancestry
