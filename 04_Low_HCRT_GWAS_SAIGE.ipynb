```bash
cd /labs/mignot/researchers/xinyuli/PCA_matching/PMRA_MIXED

awk -F'[\t ]+' -v OFS='\t' '
NR==FNR { 
    sex[$2]=$3; 
    next 
}

FNR==1 { 
    print $0, "SEX"; 
    next 
}

{ 
    s = ($2 in sex) ? sex[$2] : "NA"; 
    if (s == "0") {
        $NF = "NA";
    } 
    
    else if (s == "1") {
        s = "0";
    } 
    
    else if (s == "2") {
        s = "1";
    }
    print $0, s 
}
' /labs/mignot/researchers/xinyuli/PMRA/sex_update.txt covar_MIXED > covar_EAS_with_sex
```
